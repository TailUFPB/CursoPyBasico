##### (2.1.3.1) Números aleatórios

Um comando bastante util é o random do módulo random do NumPy, que cria um vetor com N números aleatórios entre 0 e 1

````python
random = np.random.random((5))
print(random)
````

Para valores acima de 1, simplesmente:

````python
random = 10 * np.random.random(5)
print(random)
````

Podemos também arredondar esses valores utilizando o NumPy

````python
### Alguns exemplos são o floor, ceiling e o round
### Veja a diferença

print("Valores aleatórios \n")
print(random)
print()
print("Arredondando com round  \n")
print(np.round(random))
print()
print("Adquirindo o ceiling (teto), inteiro acima do valor  \n")
print(np.ceil(random))
print()
print("Adquirindo o floor (chão), inteiro abaixo do valor  \n")
print(np.floor(random))
````

#### (2.2.5) Boolean Indexing

*   E >>> &

#### (2.3.3) Multiplos gráficos

Ao entrar num projeto onde muitos gráficos parecidos serão criados, torna-se vantajoso criar um plotter com a *def* do python

````python
### 

def plotter(ax, data1, data2, param_dict):

    graf = ax.plot(data1, data2, **param_dict)
    return graf
````

````python
data1, data2, data3, data4 = np.random.randn(4, 50)
````

````python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(5, 5)

plotter(ax, data1, data2, {'marker': 'x'})
````

````python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(11, 5)
plotter(ax1, data1, data2, {'marker': 'x'})
plotter(ax2, data3, data4, {'marker': 'o'})
````

##### (2.5.2.1) Binarizer, KBinsDiscretizer e Encoders

Encoders são uma forma de transformar categorias *string* em números inteiros, outra opção é dividí-los em colunas utilizando valores 1 e 0, isso e util por dois motivos, alguns modelos de inteligência artificial não suportam *strings* e alguns são mais eficientes quando utilizamos várias colunas com booleanos

Começando com o Binerizer, ele é um método que transforma valores numéricos em 1 e 0, colocamos um valor de corte e por padrão o que está acima é marcado como 1, abaixo é marcado como 0

````python
### Relembrando que para pegarmos a mediana usamos o mean

wine_df['hue'].mean()
````

````python
### Para esse método devemos reorganizar nosso array em 2d, o fazemos dessa forma
### .reshape(-1,1), deixando dessa forma [[a], [b], [c]]

hue_binarizer.fit(wine_df['hue'].values.reshape(-1,1))
````

Podemos ver também a matriz de confusão de uma forma bem melhor:

````python
### Esse é um plot que sempre copio e colo para visualizar melhor a confusion matrix

import itertools
plt.style.use("default")

def plot_confusion_matrix(cm, target_names, title='Confusion matrix', cmap=None, normalize=True):
    accuracy = np.trace(cm) / float(np.sum(cm))
    misclass = 1 - accuracy
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    if cmap is None:
        cmap = plt.get_cmap('Blues')
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    if target_names is not None:
        tick_marks = np.arange(len(target_names))
        plt.xticks(tick_marks, target_names, rotation=45)
        plt.yticks(tick_marks, target_names)
    thresh = cm.max() / 1.5 if normalize else cm.max() / 2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        if normalize:
            plt.text(j, i, "{:0.2f}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
        else:
            plt.text(j, i, "{:,}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label\naccuracy={:0.4f}; misclass={:0.4f}'.format(accuracy, misclass))
    plt.show()
````

````python
plot_confusion_matrix(
    metrics.confusion_matrix(y_test, modelo1.predict(X_test)), 
    ['class_0', 'class_1', 'class_2'],
    cmap = 'Blues'
    )
````

````python
plot_confusion_matrix(
    metrics.confusion_matrix(y_test, modelo2.predict(X_test)), 
    ['class_0', 'class_1', 'class_2'],
    cmap = 'Greens'
    )
````

````python
plot_confusion_matrix(
    metrics.confusion_matrix(y_test, modelo3.predict(X_test)), 
    ['class_0', 'class_1', 'class_2'],
    cmap = 'Reds'
    )
````
